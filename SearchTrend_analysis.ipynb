```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
search = pd.read_csv('data/SearchTrend.csv')
search
```

```python
search.isna().sum()
```

```python
# datetime으로 년 월 일 column 만들기
search['date'] = pd.to_datetime(search['date'])
search = search.loc[search['date'] >= "2019-06-01"]

search['year'] = search['date'].dt.year
search['month'] = search['date'].dt.month
search['day'] = search['date'].dt.day

# column 순서 바꾸기
search = search[['date', 'year', 'month', 'day', 'cold', 'flu', 'pneumonia', 'coronavirus']]
search
```

```python
# groupby로 년, 월별로 묶기
# search.groupby('year').mean()
g_search = search.groupby(['year', 'month'])[['cold', 'flu', 'pneumonia', 'coronavirus']].mean()
g_search
```

```python
# 코로나 시작 이전, 이후로 나눠서 보기
# 검색 추세 그래프로 나타내기
g_search['cold'].plot()
g_search['flu'].plot()
g_search['pneumonia'].plot()
g_search[['coronavirus']].plot()
g_search.plot(subplots=True)
```
