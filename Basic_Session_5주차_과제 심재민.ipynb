**B1.** 전국 문헌정보학과 취업률 현황.csv 를 불러와 df라는 변수에 저장하시오. (3점)

**B2.** 취업률이 높은 순서대로 정렬한 후, 취업률 상위 10개 학교명을 출력하시오. (5점)

```python
# code here
# df.sort_values('취업률', ascending=False).loc[:10, '학교명']
df.sort_values('취업률', ascending=False, ignore_index=True).loc[:10, '학교명']
```

**B3.** 취업률이 평균이상인 학교들의 졸업자수 총합을 구하시오. (7점)

```python
# code here
df.loc[df['취업률'] >= df['취업률'].mean()]['졸업자수'].sum()
```

**B4.** 반복문과 조건문을 활용하여 학제를 숫자로(대학교 : 1, 일반대학원 : 2, 전문대학 : 3) 바꾸세요. (15점)

```python
# code here
for i in range(df.shape[0]):
  if df.loc[i, '학제'] == '대학교':
    df.loc[i, '학제'] = 1
  elif df.loc[i, '학제'] == '일반대학원':
    df.loc[i, '학제'] = 2
  elif df.loc[i, '학제'] == '전문대학':
    df.loc[i, '학제'] = 3
df
```

**B5.** 반복문과 조건문을 사용하지 말고 '주야구분' 열의 값들을 다음과 같이 변경하세요. (15점)
 
'일반과정_주간' -> '주간'  
'일반과정_야간' -> '야간'  
'학사학위전공심화(경력없음)_야간' -> '야간심화'  

```python
# code here
df.loc[df['주야구분'] == '일반과정_주간', '주야구분'] = '주간'
df.loc[df['주야구분'] == '일반과정_야간', '주야구분'] = '야간'
df.loc[df['주야구분'] == '학사학위전공심화(경력없음)_야간', '주야구분'] = '야간심화'
df
```

# C.

**C1.** 출력값이 아래와 동일하도록 DataFrame을 생성하고 df라는 변수에 저장하시오. (3점)

|  | 이름	| 생년월일 | 
| ----- | ------------- | ------ |
| 0 | 남주혁 |	19940222 |
| 1 |	안효섭 | 19950417 |
| 2 | 김우빈  | 19890726|
|3 | 공유 |  19790710 |
|4 | 이준호 | 19900125|

```python
# code here
data = {
    '이름': ['남주혁', '안효섭', '김우빈', '공유', '이준호'],
    '생년월일': [19940222, 19950417, 19890726, 19790710, 19900125]
}
df = pd.DataFrame(data)
df
```

**C2.** '생년월일' 열의 속성을 datetime으로 변경해주세요. (7점)

```python
# code here
df['생년월일'] = pd.to_datetime(df['생년월일'], format='%Y%m%d')
df
```

**C3.** '생년월일' 열을 활용하여 '연도' , '월' , '일' , '나이' 라는 열을 생성해주세요. (8점)

```python
# code here
year = []
mon = []
day = []
age = []
for i in range(df.shape[0]):
  year.append(df.loc[i, '생년월일'].year)
  mon.append(df.loc[i, '생년월일'].month)
  day.append(df.loc[i, '생년월일'].day)
  age.append(pd.to_datetime('now').year - df.loc[i, '생년월일'].year + 1)

df['연도'] = year
df['월'] = mon
df['일'] = day
df['나이'] = age
df
```

**C4.** 생일이 짝수일이고 나이가 30세 이상인 사람의 이름을 출력하시오. (7점)

```python
# code here
df.query('일 % 2 == 0 and 나이 >= 30')['이름']
```
